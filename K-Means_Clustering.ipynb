```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

Mengimpor pustaka yang diperlukan untuk analisis dan visualisasi data

```python
input_data = pd.read_csv("Mall_Customers.csv")
input_data.head()
```

Membaca file dan menampilkan lima baris pertama dari data tersebut

```python
input_data.shape
```

Besar atau jumlah data set yang sedang digunakan

```python
input_data.isnull().sum()
```

Menghitung jumlah nilai null (kosong) dalam setiap kolom

```python
# selecting features
x = input_data.iloc[:,[3,4]].values
```

Memilih dua fitur tertentu dari DataFrame input_data dan mengonversinya ke dalam bentuk array numpy

```python
# Elbow Method

score = []

for cluster in range(1,11):
    kmeans = KMeans(n_clusters = cluster, init="k-means++", random_state=10)
    kmeans.fit(x)
    score.append(kmeans.inertia_)
```

Menemukan jumlah cluster yang optimal dalam algoritma K-Means

```python
# plotting the score

plt.plot(range(1,11), score)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

Membuat plot hasil dari metode siku (Elbow Method) untuk menentukan jumlah cluster yang optimal dalam algoritma K-Means

```python
# Import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

# Data yang akan di-cluster (contoh)
# Misalkan x adalah data yang akan di-cluster, dengan dua fitur
# x = np.array([[...], [...], ...])

# Loop untuk mencoba jumlah cluster dari 2 hingga 10
for n_clusters in range(2, 11):
    # Membuat subplot untuk silhouette plot dan visualisasi data terklaster
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))

    # Set batas untuk silhouette plot
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(x) + (n_clusters + 1) * 10])

    # Inisialisasi KMeans dan prediksi cluster
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(x)

    # Menghitung average silhouette score
    silhouette_avg = silhouette_score(x, cluster_labels)
    print(f"For n_clusters = {n_clusters}, The average silhouette_score is : {silhouette_avg}")

    # Menghitung silhouette score untuk setiap sample
    sample_silhouette_values = silhouette_samples(x, cluster_labels)

    # Inisialisasi nilai y_lower untuk plot
    y_lower = 10

    # Loop untuk plot setiap cluster
    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        # Pemilihan warna untuk setiap cluster
        color = cm.nipy_spectral(float(i) / n_clusters)

        # Fill plot silhouette
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label cluster di sebelah kanan silhouette plot
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Update y_lower untuk plot berikutnya
        y_lower = y_upper + 10

    # Konfigurasi plot silhouette
    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")
    ax1.set_yticks([])
    ax1.set_xticks(np.linspace(-0.1, 1, 6))

    # Visualisasi data yang sudah terklaster
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(x[:, 0], x[:, 1], marker='.', s=30, lw=0, alpha=0.7, c=colors, edgecolor='k')

    # Visualisasi pusat cluster
    centers = clusterer.cluster_centers_
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o', c="white", alpha=1, s=200, edgecolor='k')
    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker=f'${i}$', alpha=1, s=50, edgecolor='k')

    # Konfigurasi plot visualisasi data terklaster
    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    # Judul keseluruhan plot
    plt.suptitle(f"Silhouette analysis for KMeans clustering on sample data with n_clusters = {n_clusters}",
                 fontsize=14, fontweight='bold')

    # Menampilkan plot
    plt.show()
```
